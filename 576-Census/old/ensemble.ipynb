```python
import pandas as pd
import numpy as np
```

```python
clfs = ['dtc', 'kneighbor', 'lda', 'lightgbm', 'qda', 'randomforest', 'svc']
dfs = []
for clf in clfs:
    df = pd.read_csv('submit-{}.csv'.format(clf), header=None)[1]
    df.name = clf
    dfs.append(df.to_frame())
submits = pd.concat(dfs, axis=1)
submits
```

```python
submits.corr()
```

```python
submits.sum(axis=1) / len(clfs)
```

```python
sample_submit = pd.read_csv('sample_submit.csv', header=None)
sample_submit[1] = (submits.sum(axis=1) / len(clfs) > 0.5).astype(int)
sample_submit.to_csv('submit-ensemble.csv', header=False, index=False)
```
